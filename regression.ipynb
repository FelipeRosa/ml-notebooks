## How does it work?

Using linear regression, we try predicting the value of a vector of output variables $Y$ from
a vector of input variables $X$. To make such prediction we use a __linear model__

$ Y = X^T\beta $

and we find $\beta$ such that it gives us predicted values that minimize the function

$ f(X, Y, \beta) = \sum_{i=1}^N{(y_i - x_i^T\beta)^2} $.

We call $f$ the __Redisual Sum of Squares__. This is simply the sum of the differences between our predicted values and
the actual values we wanted to our predictions to be. Thus, by minimizing it, we found a linear model that minimizes that sum of squares.
